```python
client = bigquery.Client()
```

```python
date_range = ''

for i in range(1, 31):
    date_range += '_5_%d_20, ' % i
date_range += '_5_31_20'
    
QUERY = ('SELECT state, %s FROM `bigquery-public-data.covid19_usafacts.confirmed_cases` ' % date_range)

query_job = client.query(QUERY)  # API request
rows = query_job.result()  # Waits for query to finish
```

```python
raw_cases = pd.read_gbq(QUERY,
                 project_id="cmsc320finalproject",
                 dialect='standard')

raw_cases
```

```python
def sum_all_cases(row):
    return row.sum()

raw_cases['total_may_cases'] = raw_cases.drop('state', axis=1).apply(lambda row: sum_all_cases(row), axis=1)
raw_cases
```

```python
by_state = raw_cases.groupby(['state'], as_index=False).sum()

by_state
```
