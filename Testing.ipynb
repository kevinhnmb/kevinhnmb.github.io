```python
api_key = '65842CF2-80D1-4ECC-A37B-CD62E3A42E7E'

def historial_airnow_api_call(zip_code, date, distance):
    built_call = 'http://www.airnowapi.org/aq/observation/zipCode/historical/?format=application/json&zipCode=%s&date=%sT00-0000&distance=50&API_KEY=%s' % (zip_code, date, api_key)
    result = requests.get(built_call)
    
    print(result.status_code)
    
    if result.status_code == 200:
        return pd.DataFrame(result.json())
    else:
        return pd.DataFrame()
```

```python
def get_may_raw_data1():
    return_df = pd.DataFrame()
    date = '2020-%s-01'
    
    for zip_code in sample_zip_codes.values():
        for day_date in range(1,32):
            
            gen_date = date % str(day_date).rjust(2, '0')
            gen_df = historial_airnow_api_call(zip_code, gen_date, '50')
            
            return_df = return_df.append(gen_df, ignore_index=True)
    return return_df
```

```python
def historial_airnow_api_call1(zip_code, start_date, end_date, distance):
    built_call = 'http://www.airnowapi.org/aq/observation/zipCode/historical/?format=application/json&zipCode=%s&date>=%sT00-0000&date<=%sT00-0000&distance=50&API_KEY=%s' % (zip_code, start_date, end_date, api_key)
    result = requests.get(built_call)
    
    print(result.status_code)
    
    if result.status_code == 200:
        return pd.DataFrame(result.json())
    else:
        return pd.DataFrame()
```
